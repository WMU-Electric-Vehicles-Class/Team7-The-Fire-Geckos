```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.interpolate import LinearNDInterpolator
from matplotlib import pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
csv_file = ('/home/parth/Desktop/EE_project/Audi_e_Tron_Motors_ResultsForMaps.xlsx')
```

```python
df_250 = pd.read_excel(csv_file)
```

```python
df_250
```

```python
Speed = df_250['SPEED [rad/s]']
Torque = df_250['TORQUE [Nm]']
eff = df_250['EFFY [%]']
```

```python
sns.set(style = "darkgrid")

fig = plt.figure(figsize=(10,20))
ax = fig.add_subplot(111, projection = '3d')

ax.set_xlabel("Speed [rad/s]")
ax.set_ylabel("Torque [Nm] ")
ax.set_zlabel("Efficiency [%] ")

ax.plot_trisurf(Speed, Torque, eff, cmap="seismic", linewidth=0.1)

plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)

surf = ax.plot_trisurf(Speed, Torque, eff, cmap="Spectral", linewidth=0)
plt.xlabel('Speed [rad/s]')
plt.ylabel('Torque [Nm]')
```
